## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 101)
y = 2 * x

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='blue')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')

plt.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 101)
y = 2 * x
z = x**2

fig = plt.figure()

main_ax = fig.add_axes([0, 0, 1, 1])

inset_ax = fig.add_axes([0.2, 0.5, 0.4, 0.4])

main_ax.plot(x, z, color='blue')
main_ax.set_xlabel('X')
main_ax.set_ylabel('Z')

inset_ax.plot(x, y, color='blue')
inset_ax.set_title('zoom')

inset_ax.set_xlim(20, 22)
inset_ax.set_ylim(30, 50)

inset_ax.set_xlabel('X')
inset_ax.set_ylabel('Y')

plt.show()
```

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 101)
y = 2 * x
z = x**2

fig, axes = plt.subplots(nrows=1, ncols=2)

plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 101)
y = 2 * x
z = x**2

fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, linestyle='--', linewidth=2)

axes[1].plot(x, z, color='red', linewidth=2)

plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 101)
y = 2 * x
z = x**2

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

axes[0].plot(x, y, linewidth=2)

axes[1].plot(x, z, linestyle='--', linewidth=3, color='red')

plt.tight_layout()
plt.show()
```
